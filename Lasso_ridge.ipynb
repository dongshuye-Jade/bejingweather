```python
# 导入各种包
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import Ridge,RidgeCV
from sklearn.linear_model import Lasso,LassoCV
from sklearn.metrics import mean_squared_error
```

```python
# 读取数据
df = pd.read_csv('D:\\Python_lover\\weather\\2016_2017_weather_n.csv',encoding='gbk')
df.drop('date',axis=1,inplace=True)
```

```python
df.columns
```

```python
y = df['zhishu']
X = df.drop(['zhishu','wuran','Monday'],axis=1)
```

```python
# 将数据集拆分为训练集和测试集
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.7,random_state = 1234)
```

```python
# 基于可视化选择lambda
# 通过不确定的alphas值，生成不同的岭回归模型
alphas = 10 ** np.linspace(-3,3,100)
ridge_cofficients = []
for alpha in alphas:
    ridge = Ridge(alpha = alpha, normalize=True)
    ridge.fit(X_train,y_train)
    ridge_cofficients.append(ridge.coef_)

    
    
# 绘制alpha的对数与回归系数的关系
# 中文乱码和坐标轴负号的处理
plt.rcParams['font.sans-serif'] =['Microsoft YaHei']
plt.rcParams['axes.unicode_minus'] =False

#设置绘图风格
plt.style.use('ggplot')
plt.plot(alphas,ridge_cofficients)
plt.xscale('log')
plt.axis('tight')
plt.title('The relationship of alpha and Ridge')
plt.xlabel('Log Alpha')
plt.ylabel('Cofficients')
plt.show()
```

```python
# 基于CV选择lambda的值，，采用交叉验证的方法
ridge_cv = RidgeCV(alphas = alphas, normalize=True,scoring='mean_squared_error',cv= 10)
ridge_cv.fit(X_train,y_train)
```

```python
# 取出最佳的lambda的值
ridge_best_alpha = ridge_cv.alpha_
ridge_best_alpha
```

```python
# 基于最佳的lambda值的建模
ridge = Ridge(alpha=ridge_best_alpha,normalize=True)
ridge.fit(X_train,y_train)
# 岭回归系数
ridge.coef_
```

```python
# 预测
ridge_predict = ridge.predict(X_test)
#预测效果验证
RMSE= np.sqrt(mean_squared_error(y_test,ridge_predict))
RMSE
```

# LASSO回归

```python
# 通过不确定的alphas的值，生成不同的alphas的模型
alphas = 10 ** np.linspace(-3,3,100)
lasso_cofficients = []

for alpha in alphas:
    lasso = Lasso(alpha=alpha,normalize=True)
    lasso.fit(X_train,y_train)
    lasso_cofficients.append(lasso.coef_)
```

```python
    
# 绘制alpha的对数与回归系数的关系
# 中文乱码和坐标轴负号的处理
plt.rcParams['font.sans-serif'] =['Microsoft YaHei']
plt.rcParams['axes.unicode_minus'] =False

#设置绘图风格
plt.style.use('ggplot')
plt.plot(alphas,ridge_cofficients)
plt.xscale('log')
plt.axis('tight')
plt.title('alpha系数与Lasso回归系数的关系')
plt.xlabel('Log Alpha')
plt.ylabel('Cofficients')
plt.show()
```

```python
# 基于CV确定lambda的值
lasso_cv = LassoCV(alphas=alphas,normalize=True,cv=9, max_iter=10000)
lasso_cv.fit(X_train,y_train)
# 取出最佳的lambda
lasso_best_alphas = lasso_cv.alpha_
lasso_best_alphas
```

```python
# 基于最佳的lambda值建模
lasso = Lasso(alpha=lasso_best_alphas,normalize=True,max_iter=10000)
lasso.fit(X_train,y_train)
```

```python
lasso.coef_
```

```python
#预测
lasso_predict = lasso.predict(X_test)
# 预测效果验证
RMSE = np.sqrt(mean_squared_error(y_test,lasso_predict))
RMSE
```
